```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller

# Assuming 'walmart_data' is your DataFrame with the stock price data
# Make sure your DataFrame has a datetime index

# Function for Augmented Dickey-Fuller Test
def adf_test(timeseries):
    result = adfuller(timeseries, autolag='AIC')
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    print('Critical Values:', result[4])

    # Interpret the results
    if result[1] <= 0.05:
        print("Conclusion: Reject the null hypothesis. The time series is stationary.")
    else:
        print("Conclusion: Fail to reject the null hypothesis. The time series is non-stationary.")

# Perform Augmented Dickey-Fuller Test
adf_test_result = adf_test(walmart_data['Close'])

# Visualize the original time series
plt.plot(walmart_data['Close'])
plt.title('Walmart Inc. (WMT) Stock Price')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.show()
```

```python
# Findian best ARIMA Model

import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# Extract the relevant column (e.g., 'Close' prices)
closing_prices = walmart_data['Close']

# Define the split point (e.g., 80% for training, 20% for testing)
split_percentage = 0.8
split_index = int(len(closing_prices) * split_percentage)

# Split the data into training and testing sets
train_data = closing_prices[:split_index]
test_data = closing_prices[split_index:]

# Define a range of ARIMA parameters to try
p_values = range(1, 3)  # Adjust the range based on your experimentation
d_values = range(1, 2)
q_values = range(1, 3)

best_mse = float('inf')  # Initialize with a high value
best_arima_model = None
best_order = None

# Iterate over different parameter values
for p in p_values:
    for d in d_values:
        for q in q_values:
            # Train the ARIMA model on the training set
            arima_model = ARIMA(train_data, order=(p, d, q))
            arima_result = arima_model.fit()

            # Calculate Mean Squared Error (MSE) on the training set
            mse = mean_squared_error(train_data, arima_result.fittedvalues)

            # Update the best model if the current one has lower MSE
            if mse < best_mse:
                best_mse = mse
                best_arima_model = arima_result
                best_order = (p, d, q)

# Display the best model summary
print(f"Best ARIMA Model Order: {best_order}")
print(best_arima_model.summary())

```

```python
# Best Arima Model

import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# Assuming 'walmart_data' is your DataFrame containing the stock price data

# Extract the relevant column (e.g., 'Close' prices)
closing_prices = walmart_data['Close']

# Define the split point (e.g., 80% for training, 20% for testing)
split_percentage = 0.8
split_index = int(len(closing_prices) * split_percentage)

# Split the data into training and testing sets
train_data = closing_prices[:split_index]
test_data = closing_prices[split_index:]

# Best ARIMA Model Order
best_order = (2, 1, 2)

# Build the final ARIMA model with the best order using only the training data
final_arima_model = ARIMA(train_data, order=best_order)
final_arima_result = final_arima_model.fit()

# Display the final model summary
print("Final ARIMA Model Summary:")
print(final_arima_result.summary())

# Predict using the final model on the test set
arima_predictions = final_arima_result.get_forecast(steps=len(test_data))

# Calculate Mean Squared Error (MSE) on the test set
mse = mean_squared_error(test_data, arima_predictions.predicted_mean)
print(f'\nMean Squared Error (MSE) on Test Set: {mse}')

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Assuming 'walmart_data' is your DataFrame with the stock price data
# Ensure your data is in a datetime format and sorted by date

# Function for ACF and PACF plot
def plot_acf_pacf(series, lags=30):
    fig, ax = plt.subplots(1, 2, figsize=(12, 4))
    plot_acf(series, lags=lags, ax=ax[0], title='Autocorrelation Function (ACF)')
    plot_pacf(series, lags=lags, ax=ax[1], title='Partial Autocorrelation Function (PACF)')
    plt.show()

# Function for ARIMA model training and residuals analysis
def train_arima_model(data, order):
    # Splitting data into training and testing sets
    train_size = int(len(data) * 0.8)
    train, test = data[:train_size], data[train_size:]

    # Training ARIMA model
    model = ARIMA(train, order=order)
    fitted_model = model.fit()

    # Residuals analysis
    residuals = train - fitted_model.fittedvalues
    plot_acf_pacf(residuals)
    
    # Additional analysis or diagnostics can be added as needed
    
    return fitted_model

# Assuming 'walmart_data' contains the 'Close' prices
close_prices = walmart_data['Close']

# Define ARIMA order (p, d, q)
arima_order = (2, 1, 2)  # Example values, adjust based on analysis

# Train ARIMA model
arima_model = train_arima_model(close_prices, arima_order)
```

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Assuming you already have the ARIMA model trained and stored in 'arima_model'
# Replace 'arima_model' with the actual variable name of your ARIMA model

# 8.1.3 Diagnostics for ARIMA Model

# Residuals
residuals = arima_model.resid

# ACF and PACF plots of residuals
fig, ax = plt.subplots(figsize=(12, 6))
sm.graphics.tsa.plot_acf(residuals, lags=20, ax=ax, title='ACF of Residuals')
plt.show()

fig, ax = plt.subplots(figsize=(12, 6))
sm.graphics.tsa.plot_pacf(residuals, lags=20, ax=ax, title='PACF of Residuals')
plt.show()

# Normality and Independence Check
sm.qqplot(residuals, line='s')
plt.title('Q-Q Plot of Residuals')
plt.show()

# Ljung-Box test for autocorrelation in residuals
lb_test_stat, lb_test_p_value = sm.stats.acorr_ljungbox(residuals, lags=[10])
print(f"Ljung-Box Test Statistics: {lb_test_stat}")
print(f"P-values: {lb_test_p_value}")

# Check for normality using the Jarque-Bera test
jb_test_stat, jb_test_p_value, _, _ = sm.stats.jarque_bera(residuals)
print(f"Jarque-Bera Test Statistics: {jb_test_stat}")
print(f"P-value: {jb_test_p_value}")
```
